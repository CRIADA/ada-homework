```python
# According to the table we can see that the first row indicates the combined information about gun-deaths
# Reproduction of the first statement

print('Average number of deaths per year is: {}'.format(data.loc[1, 'Deaths']))
print('Average population is: {}'.format(data.loc[1, 'Population']))
print('Average death rate is: {}'.format(data.loc[1, 'Rate']))

print('\nHence according to the data - More than 33,000 people are fatally shot in the U.S. each year.')
```

```python
# Reproduction of the third, fourth, fifth statement

num_deaths = data.loc[1, 'Deaths']

data_suicide = data[data['Intent'] == 'Suicide']
#data_suicide.head()

# The first row (index=109) of the data_suicide table is the summed information about all suicide deaths
num_suicide_deaths = data_suicide.loc[109, 'Deaths']
suicide_percent = num_suicide_deaths/num_deaths*100
print('The suicide gun deaths is {} % out of all which is close to two-third.\n'.format(suicide_percent))

data_suicide_male = data_suicide[data_suicide['Gender'] == 'Male']
#data_suicide_male.head()

# The first row (index=181) of the data_suicide_male table is the summed information about all suicide male deaths
num_suicide_male_deaths = data_suicide_male.loc[181, 'Deaths']
suicide_male_percent = num_suicide_male_deaths/num_suicide_deaths*100
print('{} % of the victims are male (close to 85%).\n'.format(suicide_male_percent))

data_suicide_male_old = data_suicide_male[(data_suicide_male['Age'] == '35 - 64') |
                                          (data_suicide_male['Age'] == '65+')]
#data_suicide_male_old

# The row with index=199 indicates all male suicides with age 35 - 64
# The row with index=205 indicates all male suicides with age 65+
# For fifth statement we have men who are 35 and older
num_suicide_male_35_64_deaths = data_suicide_male_old.loc[199, 'Deaths']
num_suicide_male_65_deaths = data_suicide_male_old.loc[205, 'Deaths']
suicide_male_old_percent = (num_suicide_male_35_64_deaths + num_suicide_male_65_deaths)/num_suicide_deaths*100
print('{} % are men age of 35 and older (more than a half).'.format(suicide_male_old_percent))
```

```python
# Reproduction of the sixth, seventh statement

data_non_suicide = data[data['Intent'] != 'Suicide']
data_homicide = data[data['Intent'] == 'Homicide']

# The row with index=217 indicates all male suicides with age 35 - 64
num_homicide_deaths = data_homicide.loc[217, 'Deaths']
homicide_percent = num_homicide_deaths/num_deaths*100
print('{} % - {} of the deaths are homicidal (1/3 of all deaths)'.format(homicide_percent, num_homicide_deaths))

data_homicide_young = data_homicide[(data_homicide['Age'] == 'Under 15') | (data_homicide['Age'] == '5')]

# The row with index=223 indicates all deaths under 15
# The row with 247 indicates all deaths with Age 5
# Young men are men who are under 34
data_homicide_young_male = data_homicide[(data_homicide['Age'] == '15 - 34') |
                                    (data_homicide['Age'] == 'Under 15') | 
                                    (data_homicide['Age'] == '5') |
                                    (data_homicide['Gender'] == 'Male')]
num_homicide_young_male_deaths = data_homicide_young_male.loc[223, 'Deaths'] + \
                                 data_homicide_young_male.loc[247, 'Deaths'] + \
                                 data_homicide_young_male.loc[229, 'Deaths']
homicide_young_male_percent = num_homicide_young_male_deaths/num_homicide_deaths*100
print('{} % of the homicidal deaths are young men'.format(homicide_young_male_percent))
```
