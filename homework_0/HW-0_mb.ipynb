```python
'''Import libraries'''
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_file = "interactive_data.csv"
df = pd.read_csv(data_file)
```

```python
df_suicides = df[df["Intent"] == "Suicide"]
tot_suicides = df_suicides.Deaths.max()
tot_deaths = df.Deaths.max()
ratio_suicides = tot_suicides/tot_deaths
ratio_suicides
```

```python
df_s = df_suicides
```

```python
for name in df_s.columns:
    df_s = df_s.drop(df_s[df_s[name] == "None selected"].index)
```

```python
#Drop un-used columns and values
df_s = df_s.drop("Population", axis=1)
df_s = df_s.drop("Rate", axis=1)
df_s = df_s.drop("Unnamed: 0", axis=1)
df_s = df_s.drop(df_s[df_s["Deaths"]==0].index)
```

```python
# Some of the ages are written as 5 instead of under 15
df_s['Age'].replace(['5'], '0-14', inplace=True)
df_s['Age'].replace(['Under 15'], '0-14', inplace=True)
df_s = df_s.reset_index(drop=True)
```

```python
df_s
```

```python
df = pd.pivot_table(df_s, index=["Race", "Gender"], values=['Deaths'], columns=['Age'])
```

```python
df.reset_index(inplace=True)
```

```python
df.rename(columns={'Under_15':'0-14', '15 - 34':'15-34', '35 - 64':'35-64'}, inplace = True)
```

```python
flattened = pd.DataFrame(df.to_records())
```

```python
flattened.rename(columns={"('Race', '')":'Race', "('Gender', '')":'Gender',"('Deaths', '15-34')":"15-34",
                         "('Deaths', '35-64')":'35-64', "('Deaths', '65+')":'65+', "('Deaths', '0-14')":'0-14'},
                inplace=True)
```
